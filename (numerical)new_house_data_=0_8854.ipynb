```python
# Function to calculate outliers using IQR
def count_outliers_iqr(df, column):
    Q1 = house[column].quantile(0.25)
    Q3 = house[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = house[(house[column] < lower_bound) | (house[column] > upper_bound)]
    return len(outliers)

# Count outliers for 'REG_FEE' and 'SALES_PRICE'
outliers_Price = count_outliers_iqr(house, 'Price')

print(f"Number of outliers in 'Price': {outliers_Price}")
```

```python
# ============================================
# CatBoost Regressor â€” RMSE-focused pipeline
# Residual modeling (log-ratio to group median) + leak-safe target encoding
# ============================================

!pip -q install catboost==1.2.5

import os, math, warnings, numpy as np, pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score, mean_squared_error
from catboost import CatBoostRegressor, Pool

warnings.filterwarnings("ignore")

# ------------------ Config ------------------
CSV_PATH  = "/content/House Price India.csv"   # <-- change if needed
TEST_SIZE = 0.20
SEED      = 42
TASK_TYPE = None       # set to "GPU" if Colab GPU is enabled
HC_THRESH = 30         # high-cardinality threshold for TE
SMOOTH    = 60         # TE smoothing
CLIP_PCT  = (0.01, 0.99)  # target clipping
ITER_CAP  = 5000

# ------------------ Utils ------------------
def rmse(y, yhat):
    return math.sqrt(mean_squared_error(y, yhat))

def add_feats(X):
    X = X.copy()
    def sdiv(a, b):
        return a / np.where((b==0)|(pd.isna(b)), np.nan, b)
    # Common housing features if present
    if {"Area","Bedrooms"}.issubset(X.columns):
        X["Area_per_Bedroom"] = sdiv(X["Area"], X["Bedrooms"] + 1)
    if {"Floor No","Total Floors"}.issubset(X.columns):
        X["Floor_Ratio"] = sdiv(X["Floor No"], X["Total Floors"])
    if {"Area","Floor No"}.issubset(X.columns):
        X["Area*Floor"] = X["Area"] * X["Floor No"]
    if {"Bathrooms","Bedrooms"}.issubset(X.columns):
        X["Rooms_per_Bath"] = sdiv(X["Bedrooms"], X["Bathrooms"] + 1)
    return X

def kfold_target_encode(X_tr, y_tr, X_te, cols, n_splits=5, smoothing=SMOOTH, seed=SEED):
    """
    Out-of-fold target mean encoding (NO leakage).
    y_tr: the TRAIN target you are modeling (here we'll encode vs log-ratio residual target)
    """
    X_tr_enc, X_te_enc = X_tr.copy(), X_te.copy()
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)
    global_mean = y_tr.mean()
    for col in cols:
        oof = pd.Series(np.zeros(len(X_tr)), index=X_tr.index, dtype=float)
        for tr_idx, va_idx in kf.split(X_tr):
            tr_fold, va_fold = X_tr.iloc[tr_idx], X_tr.iloc[va_idx]
            y_tr_fold = y_tr.iloc[tr_idx]
            cnt = tr_fold[col].value_counts()
            mu  = tr_fold.join(y_tr_fold.rename("yt")).groupby(col)["yt"].mean()
            smooth = (mu * cnt + global_mean * smoothing) / (cnt + smoothing)
            oof.iloc[va_idx] = va_fold[col].map(smooth).fillna(global_mean).values

        X_tr_enc[col+"_te"] = oof.values

        cnt_full = X_tr[col].value_counts()
        mu_full  = X_tr.join(y_tr.rename("yt")).groupby(col)["yt"].mean()
        smooth_full = (mu_full * cnt_full + global_mean * smoothing) / (cnt_full + smoothing)
        X_te_enc[col+"_te"] = X_te[col].map(smooth_full).fillna(global_mean).values

    return X_tr_enc, X_te_enc

def fit_catboost(X_tr, y_tr, X_va, y_va, cat_cols):
    params = dict(
        iterations=ITER_CAP,
        learning_rate=0.03,
        depth=10,
        l2_leaf_reg=5,
        subsample=0.85,
        rsm=0.9,
        loss_function="RMSE",
        eval_metric="RMSE",
        random_seed=SEED,
        od_type="Iter",     # early stopping
        od_wait=200,
        verbose=200,
        thread_count=-1
    )
    if TASK_TYPE:
        params["task_type"] = TASK_TYPE

    model = CatBoostRegressor(**params)
    pool_tr = Pool(X_tr, y_tr, cat_features=cat_cols if len(cat_cols) else None)
    pool_va = Pool(X_va, y_va, cat_features=cat_cols if len(cat_cols) else None)
    model.fit(pool_tr, eval_set=pool_va, use_best_model=True, verbose=200)
    return model

# ------------------ Load & Base Clean ------------------
assert os.path.exists(CSV_PATH), f"File not found: {CSV_PATH}"
df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()
print("Loaded:", df.shape)

assert "Price" in df.columns, "No 'Price' column found."

# Drop leakage/useless columns
drop_cols = [c for c in df.columns if ("sqft" in c.lower() and "price" in c.lower()) or ("price per" in c.lower())]
drop_cols += [c for c in df.columns if c.lower().startswith("unnamed")]
drop_cols += [c for c in ["ID","Id","id"] if c in df.columns]
df = df.drop(columns=drop_cols, errors="ignore")

# Remove duplicates, finite target, clip severe outliers
df = df.drop_duplicates()
df = df[np.isfinite(df["Price"])]
lo, hi = df["Price"].quantile(CLIP_PCT)
df = df[(df["Price"]>=lo) & (df["Price"]<=hi)].reset_index(drop=True)

# Parse obvious date columns (expand then drop original)
for c in df.select_dtypes(include=["object"]).columns.tolist():
    try:
        dt = pd.to_datetime(df[c], errors="raise", infer_datetime_format=True)
        df[c+"_year"]  = dt.dt.year
        df[c+"_month"] = dt.dt.month
        df[c+"_dow"]   = dt.dt.dayofweek
        df.drop(columns=[c], inplace=True)
    except Exception:
        pass

# Impute missing
for c in df.select_dtypes(include="number").columns:
    df[c] = df[c].fillna(df[c].median())
for c in df.select_dtypes(include=["object","category"]).columns:
    df[c] = df[c].fillna("Missing")

# ------------------ Train/Test split ------------------
train_df, test_df = train_test_split(df, test_size=TEST_SIZE, random_state=SEED, shuffle=True)

# Keys for grouping signal (use what exists)
KEYS = [k for k in ["City","Area Name","Location","Locality","Type of Property","Property Type"] if k in train_df.columns]

# If none exist, make a coarse key from (maybe) pin/facing/society if present
if not KEYS:
    for c in ["Society","Facing","Developer"]:
        if c in train_df.columns:
            KEYS.append(c)
if not KEYS:
    # final fallback: single key to avoid errors, will use global median
    KEYS = ["__GLOBAL__"]
    train_df["__GLOBAL__"] = "ALL"
    test_df["__GLOBAL__"]  = "ALL"

# ------------------ Group medians (train-only!) ------------------
train_group_median = train_df.groupby(KEYS)["Price"].median()
train_global_median = train_df["Price"].median()

def attach_group_median(df_part):
    m = df_part[KEYS].merge(
        train_group_median.rename("grp_med"),
        left_on=KEYS, right_index=True, how="left"
    )["grp_med"].fillna(train_global_median).values
    return m

train_grp_med = attach_group_median(train_df)
test_grp_med  = attach_group_median(test_df)

# ------------------ Targets: log-ratio residual ------------------
# y* = log(Price+1) - log(group_median+1)
y_train_star = np.log1p(train_df["Price"].values) - np.log1p(train_grp_med)
y_test_star  = np.log1p(test_df["Price"].values)  - np.log1p(test_grp_med)

# ------------------ Features ------------------
X_train = train_df.drop(columns=["Price"]).copy()
X_test  = test_df.drop(columns=["Price"]).copy()

# Add engineered features
X_train = add_feats(X_train)
X_test  = add_feats(X_test)

# High-cardinality categoricals -> target encode against y_star (NO leakage)
high_card = [c for c in X_train.select_dtypes(include=["object","category"]).columns
             if X_train[c].nunique() > HC_THRESH]

if high_card:
    X_train, X_test = kfold_target_encode(X_train, pd.Series(y_train_star, index=X_train.index),
                                          X_test, high_card, n_splits=5, smoothing=SMOOTH, seed=SEED)
    # Drop original high-card to reduce overfit
    X_train = X_train.drop(columns=high_card)
    X_test  = X_test.drop(columns=high_card)

# Low-card categoricals for CatBoost native handling
cat_cols = X_train.select_dtypes(include=["object","category"]).columns.tolist()

# Internal validation split for early stopping
X_tr, X_va, y_tr, y_va = train_test_split(X_train, y_train_star, test_size=0.2, random_state=SEED)

# ------------------ Train ------------------
model = fit_catboost(X_tr, y_tr, X_va, y_va, cat_cols)

# ------------------ Predict in log-ratio space & reconstruct ------------------
y_star_pred_test = model.predict(X_test)
# reconstruct predicted log(price+1) = log(grp_med+1) + y_star_pred
pred_log_price = np.log1p(test_grp_med) + y_star_pred_test
y_pred = np.expm1(pred_log_price)

# ------------------ Metrics ------------------
r2   = r2_score(test_df["Price"].values, y_pred)
err  = rmse(test_df["Price"].values, y_pred)

print("\n=== Held-out Test Metrics ===")
print(f"RÂ²:            {r2:.4f}")
print(f"RMSE (â‚¹):      {err:,.0f}")
print(f"RMSE (â‚¹ lakh): {err/1e5:,.2f}")

# ------------------ Save comparison ------------------
cmp = pd.DataFrame({
    "Actual_Price": test_df["Price"].values,
    "Group_Median": test_grp_med,
    "Predicted_Price": y_pred,
    "Abs_Error": np.abs(test_df["Price"].values - y_pred)
})
cmp.to_csv("catboost_residual_comparison.csv", index=False)
print("\nSaved: catboost_residual_comparison.csv")

# ------------------ Notes to squeeze further ------------------
# - Enable GPU: set TASK_TYPE = "GPU"
# - Try depth=12 and iterations up to 8000 (OD will stop early)
# - Add more location keys (PinCode/lat-long if available)
# - Add age_of_property, amenity flags (parking/lift/gated), time (year/quarter)
# - Increase SMOOTH for TE if high-card columns are noisy (e.g., 100â€“150)
```

```python
# ================================
# CatBoost on log(Price) + group-median feature (no residual target)
# ================================
!pip -q install catboost==1.2.5

import os, math, warnings, numpy as np, pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score, mean_squared_error
from catboost import CatBoostRegressor, Pool
warnings.filterwarnings("ignore")

# ---------- CONFIG ----------
CSV_PATH  = "/content/House Price India.csv"  # change if needed
SEED      = 42
TEST_SIZE = 0.20
TASK_TYPE = None   # set "GPU" if GPU available
HC_THRESH = 30
SMOOTH    = 80
CLIP_PCT  = (0.005, 0.995)   # softer clipping than before
ITER_CAP  = 8000

# ---------- Utils ----------
def rmse(y, yhat): return math.sqrt(mean_squared_error(y, yhat))

def add_feats(X):
    X = X.copy()
    def sdiv(a,b): return a/np.where((b==0)|(pd.isna(b)), np.nan, b)
    if {"Area","Bedrooms"}.issubset(X.columns):
        X["Area_per_Bedroom"] = sdiv(X["Area"], X["Bedrooms"]+1)
    if {"Floor No","Total Floors"}.issubset(X.columns):
        X["Floor_Ratio"] = sdiv(X["Floor No"], X["Total Floors"])
    if {"Area","Floor No"}.issubset(X.columns):
        X["Area*Floor"] = X["Area"] * X["Floor No"]
    if {"Bathrooms","Bedrooms"}.issubset(X.columns):
        X["Rooms_per_Bath"] = sdiv(X["Bedrooms"], X["Bathrooms"]+1)
    return X

def kfold_te(X_tr, y_tr, X_te, cols, n_splits=5, smoothing=SMOOTH):
    Xtr, Xte = X_tr.copy(), X_te.copy()
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=SEED)
    gmean = y_tr.mean()
    for col in cols:
        oof = pd.Series(np.zeros(len(Xtr)), index=Xtr.index)
        for tr_idx, va_idx in kf.split(Xtr):
            trf, vaf = Xtr.iloc[tr_idx], Xtr.iloc[va_idx]
            yf = y_tr.iloc[tr_idx]
            cnt = trf[col].value_counts()
            mu  = trf.join(yf.rename("y")).groupby(col)["y"].mean()
            smooth = (mu*cnt + gmean*smoothing)/(cnt+smoothing)
            oof.iloc[va_idx] = vaf[col].map(smooth).fillna(gmean).values
        Xtr[col+"_te"] = oof.values
        cntF = Xtr[col].value_counts()
        muF  = Xtr.join(y_tr.rename("y")).groupby(col)["y"].mean()
        smoothF = (muF*cntF + gmean*smoothing)/(cntF+smoothing)
        Xte[col+"_te"] = Xte[col].map(smoothF).fillna(gmean).values
    return Xtr, Xte

def fit_cat(Xtr, ytr, Xva, yva, cat_cols):
    params = dict(
        iterations=ITER_CAP,
        learning_rate=0.02,   # a bit slower, better generalization
        depth=10,
        l2_leaf_reg=6,
        subsample=0.9,
        rsm=0.9,
        loss_function="RMSE",
        eval_metric="RMSE",
        random_seed=SEED,
        od_type="Iter",
        od_wait=300,          # give it more patience
        verbose=200,
        thread_count=-1
    )
    if TASK_TYPE: params["task_type"] = TASK_TYPE
    m = CatBoostRegressor(**params)
    pool_tr = Pool(Xtr, ytr, cat_features=cat_cols if len(cat_cols) else None)
    pool_va = Pool(Xva, yva, cat_features=cat_cols if len(cat_cols) else None)
    m.fit(pool_tr, eval_set=pool_va, use_best_model=True, verbose=200)
    return m

# ---------- Load & prep ----------
assert os.path.exists(CSV_PATH), "CSV not found"
df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()
assert "Price" in df.columns, "No 'Price' column"

# remove leakage/useless
drop_cols = [c for c in df.columns if ("sqft" in c.lower() and "price" in c.lower()) or ("price per" in c.lower())]
drop_cols += [c for c in df.columns if c.lower().startswith("unnamed")]
drop_cols += [c for c in ["ID","Id","id"] if c in df.columns]
df = df.drop(columns=drop_cols, errors="ignore")
df = df.drop_duplicates()
df = df[np.isfinite(df["Price"])]

# gentle target clip (helps RMSE without losing too much)
lo, hi = df["Price"].quantile(CLIP_PCT)
df = df[(df["Price"]>=lo) & (df["Price"]<=hi)].reset_index(drop=True)

# parse any date-like object cols to y/m/dow features
for c in df.select_dtypes(include=["object"]).columns.tolist():
    try:
        dt = pd.to_datetime(df[c], errors="raise", infer_datetime_format=True)
        df[c+"_year"] = dt.dt.year; df[c+"_month"] = dt.dt.month; df[c+"_dow"] = dt.dt.dayofweek
        df.drop(columns=[c], inplace=True)
    except Exception:
        pass

# impute missing
for c in df.select_dtypes(include="number").columns:
    df[c] = df[c].fillna(df[c].median())
for c in df.select_dtypes(include=["object","category"]).columns:
    df[c] = df[c].fillna("Missing")

# split
train_df, test_df = train_test_split(df, test_size=TEST_SIZE, random_state=SEED, shuffle=True)

# keys for group median feature
KEYS = [k for k in ["City","Area Name","Location","Locality","Type of Property","Property Type"] if k in train_df.columns]
if not KEYS:
    for c in ["Society","Facing","Developer"]:
        if c in train_df.columns: KEYS.append(c)
if not KEYS:
    KEYS = ["__GLOBAL__"]; train_df["__GLOBAL__"]="ALL"; test_df["__GLOBAL__"]="ALL"

# group median (train-only) as a FEATURE
train_med = train_df.groupby(KEYS)["Price"].median()
global_med = train_df["Price"].median()

def add_grp_med(df_part):
    m = df_part[KEYS].merge(train_med.rename("grp_med"), left_on=KEYS, right_index=True, how="left")["grp_med"]
    return m.fillna(global_med).values

train_grp_med = add_grp_med(train_df)
test_grp_med  = add_grp_med(test_df)
train_df["log_grp_med"] = np.log1p(train_grp_med)
test_df["log_grp_med"]  = np.log1p(test_grp_med)

# features/target (direct log-price target)
X_train = train_df.drop(columns=["Price"]).copy()
X_test  = test_df.drop(columns=["Price"]).copy()
X_train = add_feats(X_train); X_test = add_feats(X_test)

y_train_log = np.log1p(train_df["Price"].values)
y_test_true = test_df["Price"].values

# leak-safe TE on high-card categoricals (vs y_train_log)
high_card = [c for c in X_train.select_dtypes(include=["object","category"]).columns
             if X_train[c].nunique() > HC_THRESH]
if high_card:
    X_train, X_test = kfold_te(X_train, pd.Series(y_train_log, index=X_train.index), X_test, high_card, n_splits=5, smoothing=SMOOTH)
    X_train = X_train.drop(columns=high_card); X_test = X_test.drop(columns=high_card)

# low-card cats for CatBoost
cat_cols = X_train.select_dtypes(include=["object","category"]).columns.tolist()

# internal val split
X_tr, X_va, y_tr, y_va = train_test_split(X_train, y_train_log, test_size=0.2, random_state=SEED)

# train
model = fit_cat(X_tr, y_tr, X_va, y_va, cat_cols)

# predict & back-transform
y_pred_log = model.predict(X_test)
y_pred = np.expm1(y_pred_log)

# metrics
r2 = r2_score(y_test_true, y_pred)
err = rmse(y_test_true, y_pred)
print("\n=== Held-out Test Metrics (direct log target) ===")
print(f"RÂ²:            {r2:.4f}")
print(f"RMSE (â‚¹):      {err:,.0f}")
print(f"RMSE (â‚¹ lakh): {err/1e5:,.2f}")

# save
pd.DataFrame({"Actual_Price": y_test_true, "Predicted_Price": y_pred,
              "Abs_Error": np.abs(y_test_true - y_pred)}).to_csv("catboost_directlog_comparison.csv", index=False)
print("Saved: catboost_directlog_comparison.csv")
```

```python
# ============================================
# CatBoost RMSE Squeezer â€” per-city models + TE + log target + 5-seed ensemble
# ============================================
!pip -q install catboost==1.2.5

import os, math, warnings, numpy as np, pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score, mean_squared_error
from catboost import CatBoostRegressor, Pool
warnings.filterwarnings("ignore")

# -------- Config --------
CSV_PATH  = "/content/House Price India.csv"   # <- change if needed
TEST_SIZE = 0.20
TASK_TYPE = None        # "GPU" if GPU available
HC_THRESH = 30          # high-card threshold for TE
SMOOTH    = 100         # TE smoothing (stronger to reduce noise)
CLIP_PCT  = (0.002, 0.998)  # very gentle clipping
AREA_MIN  = 200         # min rows to train per-area model
CITY_MIN  = 200         # min rows to train per-city model
SEEDS     = [11, 22, 33, 44, 55]  # ensemble seeds

# -------- Utilities --------
def rmse(y, yhat): return math.sqrt(mean_squared_error(y, yhat))

def add_feats(X):
    X = X.copy()
    def sdiv(a,b): return a/np.where((b==0)|(pd.isna(b)), np.nan, b)
    # generic engineered features
    if "Area" in X.columns:
        X["log_Area"] = np.log1p(X["Area"])
    if {"Area","Bedrooms"}.issubset(X.columns):
        X["Area_per_Bedroom"] = sdiv(X["Area"], X["Bedrooms"]+1)
    if {"Area","Bathrooms"}.issubset(X.columns):
        X["Area_per_Bath"] = sdiv(X["Area"], X["Bathrooms"]+1)
    if {"Floor No","Total Floors"}.issubset(X.columns):
        X["Floor_Ratio"] = sdiv(X["Floor No"], X["Total Floors"])
    if {"Area","Floor No"}.issubset(X.columns):
        X["Area*Floor"] = X["Area"] * X["Floor No"]
    if {"Bedrooms","Bathrooms"}.issubset(X.columns):
        X["Rooms_per_Bath"] = sdiv(X["Bedrooms"], X["Bathrooms"]+1)
    return X

def parse_dates(df):
    for c in df.select_dtypes(include=["object"]).columns.tolist():
        try:
            dt = pd.to_datetime(df[c], errors="raise", infer_datetime_format=True)
            df[c+"_year"] = dt.dt.year; df[c+"_month"] = dt.dt.month; df[c+"_dow"] = dt.dt.dayofweek
            df.drop(columns=[c], inplace=True)
        except Exception:
            pass
    return df

def kfold_target_encode(X_tr, y_tr, X_te, cols, n_splits=5, smoothing=SMOOTH, seed=42):
    Xtr, Xte = X_tr.copy(), X_te.copy()
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed)
    gmean = y_tr.mean()
    for col in cols:
        oof = pd.Series(np.zeros(len(Xtr)), index=Xtr.index, dtype=float)
        for tr_idx, va_idx in kf.split(Xtr):
            trf, vaf = Xtr.iloc[tr_idx], Xtr.iloc[va_idx]
            yf = y_tr.iloc[tr_idx]
            cnt = trf[col].value_counts()
            mu  = trf.join(yf.rename("y")).groupby(col)["y"].mean()
            smooth = (mu*cnt + gmean*smoothing)/(cnt+smoothing)
            oof.iloc[va_idx] = vaf[col].map(smooth).fillna(gmean).values
        Xtr[col+"_te"] = oof.values
        cntF = Xtr[col].value_counts()
        muF  = Xtr.join(y_tr.rename("y")).groupby(col)["y"].mean()
        smoothF = (muF*cntF + gmean*smoothing)/(cntF+smoothing)
        Xte[col+"_te"] = Xte[col].map(smoothF).fillna(gmean).values
    return Xtr, Xte

def fit_one(X_tr, y_tr, X_va, y_va, cat_cols, seed):
    # small param sweep per seed
    configs = [
        dict(iterations=6000, learning_rate=0.018, depth=10, l2_leaf_reg=6, subsample=0.9, rsm=0.9),
        dict(iterations=7000, learning_rate=0.016, depth=10, l2_leaf_reg=7, subsample=0.9, rsm=0.9),
        dict(iterations=8000, learning_rate=0.014, depth=12, l2_leaf_reg=6, subsample=0.9, rsm=0.9),
    ]
    best_model, best_rmse = None, 1e18
    pool_va = Pool(X_va, y_va, cat_features=cat_cols if len(cat_cols) else None)
    for cfg in configs:
        params = dict(
            loss_function="RMSE", eval_metric="RMSE",
            random_seed=seed, od_type="Iter", od_wait=400,
            verbose=200, thread_count=-1
        )
        if TASK_TYPE: params["task_type"] = TASK_TYPE
        params.update(cfg)
        m = CatBoostRegressor(**params)
        pool_tr = Pool(X_tr, y_tr, cat_features=cat_cols if len(cat_cols) else None)
        m.fit(pool_tr, eval_set=pool_va, use_best_model=True, verbose=200)
        pred_va = m.predict(X_va)
        e = rmse(np.expm1(y_va), np.expm1(pred_va))
        if e < best_rmse:
            best_rmse, best_model = e, m
    return best_model

def train_block(train_block, test_block, tag, keys_for_median):
    # -------- group median feature (train-only) --------
    med = train_block.groupby(keys_for_median)["Price"].median()
    global_med = train_block["Price"].median()
    def attach_med(dfpart):
        m = dfpart[keys_for_median].merge(med.rename("grp_med"), left_on=keys_for_median, right_index=True, how="left")["grp_med"]
        return m.fillna(global_med).values
    train_block = train_block.copy(); test_block = test_block.copy()
    train_block["log_grp_med"] = np.log1p(attach_med(train_block))
    test_block["log_grp_med"]  = np.log1p(attach_med(test_block))

    # -------- features/target (direct log target) --------
    X_tr_full = train_block.drop(columns=["Price"])
    X_te_full = test_block.drop(columns=["Price"])
    X_tr_full = add_feats(X_tr_full); X_te_full = add_feats(X_te_full)

    y_tr_full = np.log1p(train_block["Price"].values)
    y_te_true = test_block["Price"].values

    # TE on high-card categoricals
    high_card = [c for c in X_tr_full.select_dtypes(include=["object","category"]).columns
                 if X_tr_full[c].nunique() > HC_THRESH]
    if high_card:
        X_tr_full, X_te_full = kfold_target_encode(X_tr_full, pd.Series(y_tr_full, index=X_tr_full.index),
                                                   X_te_full, high_card, n_splits=5, smoothing=SMOOTH, seed=42)
        X_tr_full = X_tr_full.drop(columns=high_card); X_te_full = X_te_full.drop(columns=high_card)

    # remaining cats for CatBoost
    cat_cols = X_tr_full.select_dtypes(include=["object","category"]).columns.tolist()

    # internal val split
    X_tr, X_va, y_tr, y_va = train_test_split(X_tr_full, y_tr_full, test_size=0.2, random_state=42)

    # -------- 5-seed ensemble --------
    preds = []
    for s in SEEDS:
        model = fit_one(X_tr, y_tr, X_va, y_va, cat_cols, seed=s)
        preds.append(np.expm1(model.predict(X_te_full)))
    y_pred = np.mean(np.stack(preds, axis=0), axis=0)

    # metrics for block
    r2 = r2_score(y_te_true, y_pred)
    e  = rmse(y_te_true, y_pred)
    print(f"[{tag}] R2={r2:.4f}  RMSE=â‚¹{e:,.0f} (â‚¹{e/1e5:.2f} lakh)")
    return pd.Series(y_pred, index=test_block.index), pd.Series(y_te_true, index=test_block.index)

# -------- Load & base clean --------
assert os.path.exists(CSV_PATH), "CSV not found"
df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()
assert "Price" in df.columns, "No 'Price' column"

# de-leak
drop_cols = [c for c in df.columns if ("sqft" in c.lower() and "price" in c.lower()) or ("price per" in c.lower())]
drop_cols += [c for c in df.columns if c.lower().startswith("unnamed")]
drop_cols += [c for c in ["ID","Id","id"] if c in df.columns]
df = df.drop(columns=drop_cols, errors="ignore").drop_duplicates()
df = df[np.isfinite(df["Price"])]

# gentle clip
lo, hi = df["Price"].quantile(CLIP_PCT)
df = df[(df["Price"]>=lo) & (df["Price"]<=hi)].reset_index(drop=True)

# parse date-like
df = parse_dates(df)

# impute
for c in df.select_dtypes(include="number").columns:
    df[c] = df[c].fillna(df[c].median())
for c in df.select_dtypes(include=["object","category"]).columns:
    df[c] = df[c].fillna("Missing")

# split
train_df, test_df = train_test_split(df, test_size=TEST_SIZE, random_state=42, shuffle=True)

# grouping keys
CITY = next((c for c in ["City","city"] if c in df.columns), None)
AREA = next((c for c in ["Area Name","Locality","Location"] if c in df.columns), None)
TYPE = next((c for c in ["Type of Property","Property Type"] if c in df.columns), None)

# decide strategy: per-city, then per-area within city if enough rows
pred_all, true_all = [], []
if CITY and train_df[CITY].nunique() > 1:
    for city, te_city in test_df.groupby(CITY):
        tr_city = train_df[train_df[CITY]==city]
        if len(tr_city) < CITY_MIN:
            # fall back to global
            y_pred, y_true = train_block(train_df, te_city, f"ALL_for_{city}",
                                         [k for k in [CITY, AREA, TYPE] if k in train_df.columns])
            pred_all.append(y_pred); true_all.append(y_true); continue
        # within city: per-area if available
        if AREA and tr_city[AREA].nunique() > 1:
            for area, te_area in te_city.groupby(AREA):
                tr_area = tr_city[tr_city[AREA]==area]
                if len(tr_area) >= AREA_MIN:
                    y_pred, y_true = train_block(tr_area, te_area, f"{city}__{area}",
                                                 [k for k in [AREA, TYPE] if k in tr_area.columns] or [CITY])
                else:
                    y_pred, y_true = train_block(tr_city, te_area, f"{city}_ALL_for_{area}",
                                                 [k for k in [CITY, AREA, TYPE] if k in tr_city.columns])
                pred_all.append(y_pred); true_all.append(y_true)
        else:
            y_pred, y_true = train_block(tr_city, te_city, f"{city}",
                                         [k for k in [CITY, TYPE] if k in tr_city.columns])
            pred_all.append(y_pred); true_all.append(y_true)
else:
    # no city: single global (still uses grouping medians if any)
    keys = [k for k in [AREA, TYPE] if k] or ["__GLOBAL__"]
    if keys == ["__GLOBAL__"]:
        train_df["__GLOBAL__"]="ALL"; test_df["__GLOBAL__"]="ALL"
    y_pred, y_true = train_block(train_df, test_df, "GLOBAL", keys)
    pred_all.append(y_pred); true_all.append(y_true)

# aggregate
y_pred_full = pd.concat(pred_all).sort_index().values
y_true_full = pd.concat(true_all).sort_index().values

print("\n=== Aggregate held-out metrics ===")
R2 = r2_score(y_true_full, y_pred_full)
E  = rmse(y_true_full, y_pred_full)
print(f"RÂ²:            {R2:.4f}")
print(f"RMSE (â‚¹):      {E:,.0f}")
print(f"RMSE (â‚¹ lakh): {E/1e5:,.2f}")

# save
pd.DataFrame({"Actual_Price": y_true_full,
              "Predicted_Price": y_pred_full,
              "Abs_Error": np.abs(y_true_full - y_pred_full)}
            ).to_csv("catboost_city_area_ensemble.csv", index=False)
print("Saved: catboost_city_area_ensemble.csv")
```
